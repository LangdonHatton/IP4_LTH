```python
car_sales = pd.read_csv('car-sales.csv')
```

```python
car_sales.head()
```

```python
car_sales['Odometer (KM)'].plot()
```

```python
#the price plot doesnt work because ther eis the dollar sign so it will not read it as a number
```

```python
car_sales['Price'] = car_sales['Price'].str.replace(r'[\$\,]', '').astype('int64')
```

```python
car_sales.head()
```

```python
car_sales['Total Sales'] = car_sales['Price'].cumsum()
```

```python
car_sales.head()
```

```python
!pip install datetime
```

```python
import datetime
num_rows = len(car_sales)
today = datetime.datetime.today()
dates = pd.date_range(start=today, periods=num_rows)
car_sales['Sale Date'] = dates.date
```

```python
car_sales.head()
```

```python
line=plt.plot(car_sales['Price'], car_sales['Total Sales'])
plt.title('Price vs total sales line plot')
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.show()
```

```python
scat=plt.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
plt.title('Odometer vs Price scatter plot')
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.show()
```

```python
bar=plt.bar(car_sales['Make'], car_sales['Odometer (KM)'])
plt.title('Make vs Odometet bar plot')
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.show()
```

```python
hist=plt.hist(car_sales['Make'])
plt.title('Histogram of car make')
plt.xlabel('Make')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
grid = fig.add_gridspec(nrows=2, ncols=2)

ax1 = fig.add_subplot(grid[0, 0])
ax1.plot(car_sales['Price'], car_sales['Total Sales'])
ax1.set_title('Price vs total sales line plot')
ax1.set_xlabel('Price')
ax1.set_ylabel('Total Sales')

ax2 = fig.add_subplot(grid[0, 1])
ax2.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
ax2.set_title('Odometer vs Price Scatter plot')
ax2.set_xlabel('Odometer (KM)')
ax2.set_ylabel('Price')

ax3 = fig.add_subplot(grid[1, 0])
ax3.bar(car_sales['Make'], car_sales['Odometer (KM)'])
ax3.set_title('Make vs Odometet bar plot')
ax3.set_xlabel('Make')
ax3.set_ylabel('Odometer (KM)')

ax4 = fig.add_subplot(grid[1, 1])
ax4.hist(car_sales['Make'])
ax4.set_title('Histogram of car make')
ax4.set_xlabel('Make')

plt.show()
```

```python
sns.boxplot(x='Colour', y='Total Sales', data=car_sales)
plt.title('Box Plot of Colour vs Total Sales')
plt.xlabel('Colour')
plt.ylabel('Total Sales')
plt.show()
```

```python
sns.lmplot(y='Odometer (KM)', x='Total Sales', hue='Make', data=car_sales, ci=None)



plt.ylabel('Odometer (KM)')
plt.xlabel('Total Sales')

plt.show()
```

```python
sns.pairplot(car_sales[['Make', 'Colour', 'Odometer (KM)', 'Doors', 'Price', 'Total Sales', 'Sale Date']])
```

```python
corr_matrix = car_sales[['Total Sales', 'Price', 'Odometer (KM)']].corr(method='pearson')
sns.heatmap(corr_matrix, annot=True, cmap='Blues')
```
